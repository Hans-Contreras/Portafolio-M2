```python
def sumar(a, b):
  """Esta función suma dos números."""
  return a + b

def restar(a, b):
  """Esta función resta dos números."""
  return a - b

def multiplicar(a, b):
  """Esta función multiplica dos números."""
  return a * b

def dividir(a, b):
  """Esta función divide dos números.
  Maneja la división por cero.
  """
  if b == 0:
    return "Error: No se puede dividir por cero."
  else:
    return a / b

def mostrar_menu_operaciones():
  """Muestra el menú numérico de operaciones."""
  print("\nSelecciona una operación del menú:")
  print("1. Sumar")
  print("2. Restar")
  print("3. Multiplicar")
  print("4. Dividir")

# Función principal de la calculadora
def calculadora():
  print("Calculadora Básica en Python")
  print("Ingresa dos números y elige una operación.")
  print("La calculadora se cerrará después de realizar una operación.")
  print("-" * 30)

  # Pedir primer número con validación
  while True:
    entrada_num1 = input("Ingresa el primer número: ")
    try:
      num1 = float(entrada_num1)
      break # Sale del bucle si la conversión es exitosa
    except ValueError:
      print("Entrada inválida. Por favor, ingresa un valor numérico.")

  # Pedir segundo número con validación
  while True:
    entrada_num2 = input("Ingresa el segundo número: ")
    try:
      num2 = float(entrada_num2)
      break # Sale del bucle si la conversión es exitosa
    except ValueError:
      print("Entrada inválida. Por favor, ingresa un valor numérico.")

  # Pedir la operación y realizarla
  while True:
    operacion_elegida = input(f"¿Qué operación quieres realizar con {num1} y {num2}? (+, -, *, /, menu): ").lower()
    resultado = None
    operacion_realizada = False

    if operacion_elegida == '+':
      resultado = sumar(num1, num2)
      print(f"Resultado: {num1} + {num2} = {resultado}")
      operacion_realizada = True
    elif operacion_elegida == '-':
      resultado = restar(num1, num2)
      print(f"Resultado: {num1} - {num2} = {resultado}")
      operacion_realizada = True
    elif operacion_elegida == '*':
      resultado = multiplicar(num1, num2)
      print(f"Resultado: {num1} * {num2} = {resultado}")
      operacion_realizada = True
    elif operacion_elegida == '/':
      resultado = dividir(num1, num2)
      print(f"Resultado: {num1} / {num2} = {resultado}")
      operacion_realizada = True
    elif operacion_elegida == 'menu':
      mostrar_menu_operaciones()
      opcion_menu = input("Ingresa tu opción del menú (1/2/3/4): ")
      if opcion_menu == '1':
        resultado = sumar(num1, num2)
        print(f"Resultado: {num1} + {num2} = {resultado}")
        operacion_realizada = True
      elif opcion_menu == '2':
        resultado = restar(num1, num2)
        print(f"Resultado: {num1} - {num2} = {resultado}")
        operacion_realizada = True
      elif opcion_menu == '3':
        resultado = multiplicar(num1, num2)
        print(f"Resultado: {num1} * {num2} = {resultado}")
        operacion_realizada = True
      elif opcion_menu == '4':
        resultado = dividir(num1, num2)
        print(f"Resultado: {num1} / {num2} = {resultado}")
        operacion_realizada = True
      else:
        print("Opción del menú inválida. Intenta de nuevo.")
        # No salimos del bucle, pedimos la operación otra vez
    else:
      print("Operación no reconocida. Intenta con '+', '-', '*', '/', o 'menu'.")

    if operacion_realizada:
      break # Sale del bucle de selección de operación si se completó una

  print("-" * 30)
  print("Cálculo completado. Gracias por usar la calculadora.")

# Ejecutar la calculadora
if __name__ == "__main__":
  calculadora()



```
